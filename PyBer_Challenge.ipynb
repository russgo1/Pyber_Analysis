# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
city_rides = pyber_data_df.groupby("type")["ride_id"].count()
city_rides
```

```python
# Checking total number of drivers
# pyber_data_df["driver_count"].sum()
```

```python
# 2. Get the total drivers for each city type
# Must be calculated by city_drivers, because 
# the pyber_data_df rows have the whole driver_count repeated in every row
city_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
city_drivers
```

```python
#  3. Get the total amount of fares for each city type
city_fares = pyber_data_df.groupby("type")["fare"].sum()
city_fares
```

```python
#  4. Get the average fare per ride for each city type. 
# Per-city fare divided by sum of all fares
mean_city_fare = city_fares / (pyber_data_df.groupby("type")["ride_id"].count())
mean_city_fare
```

```python
# 5. Get the average fare per driver for each city type. 
# fares by city type divided by number of drivers per city type
mean_driver_fare = city_fares / city_drivers
mean_driver_fare
```

```python
#  6. Create a PyBer summary DataFrame. 
# List of series to be integrated into data frame
data_list = [city_rides, city_drivers, city_fares, mean_city_fare, mean_driver_fare]
# Dictionary for new column names
column_rename_dict = {"ride_id":"Total Rides",
                      "driver_count":"Total Drivers", 
                      "fare":"Total Fares", 
                      0:"Average Fare per Ride", 
                      1:"Average Fare per Driver"}
# Build DataFrame
pyber_summary_df = pd.concat(data_list, axis=1)
# Rename columns
pyber_summary_df.rename(columns=column_rename_dict, inplace=True)

# Display DataFrame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
# Total fares
pyber_summary_df.iloc[:,2] = pyber_summary_df.iloc[:,2].map('${:.2f}'.format)
# Avg fare per ride
pyber_summary_df.iloc[:,3] = pyber_summary_df.iloc[:,3].map('${:.2f}'.format)
# Avg fare per driver
pyber_summary_df.iloc[:,4] =pyber_summary_df.iloc[:,4].map('${:.2f}'.format)

# Display DF
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
type_date_df = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
type_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
type_date_df = type_date_df.reset_index()
type_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
type_data_pivot = type_date_df.pivot(index='date',
                                    columns='type',
                                    values='fare')

type_data_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
Jan_April_2019_df = type_data_pivot.loc["2019-01-01":"2019-04-29"]
Jan_April_2019_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
Jan_April_2019_df.index = pd.to_datetime(Jan_April_2019_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
Jan_April_2019_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_april_2019_weeks_df = Jan_April_2019_df.resample('W').sum()
jan_april_2019_weeks_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Create Line Chart
jan_april_2019_weeks_df.plot(figsize=(25,9))

# Chart titles and formatting
plt.title("Total Fare by City Type", fontsize=20)
plt.xlabel("Month (2019)", fontsize=14)
plt.ylabel("Fare ($USD)", fontsize=14)

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Save chart
plt.savefig("analysis/Pyber_fare_summary.png")
```
