## 5.1.3

```python
# Set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Creat plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Alternate method with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Quick plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create plot and add lable for legend
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set y limits between 0 and 45
plt.ylim(0, 45)
# Create title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

### Object-Oriented Approach

```python
# Create the plot, plot size, and line label, color and marker
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x_axis, y_axis, color='green', marker='D', linewidth='2', label='Boston')
# Axes limits
ax.set_xlim("Jan", "June")
ax.set_ylim(0, 45)
# Axes labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Title
ax.set_title("PyBer Fares by Month")
# Add grid
ax.grid()
# Add legend
ax.legend()
```

## 5.1.5 Bar Charts

```python
# Set x-axis to the months
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set y-axis to fares collected for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create chart
plt.bar(x_axis, y_axis)
```

```python
# Add annotations
```

```python
# Create plot with color and label
plt.bar(x_axis, y_axis, color='green', label='Boston')
# Axis limits
plt.ylim(0, 45)
# Axis labels
plt.xlabel("Month")
plt.ylabel("Fare ($)")
# Title
plt.title("Pyber Fare by Month")
# Addthe legend
plt.legend()
```

### Horizontal Bar Chart

```python
# Create chart
plt.barh(x_axis, y_axis)
```

```python
# Switch the axes
plt.barh(y_axis, x_axis)
```

```python
# gca stands for get current axis
# used to perform functions on particular axis

# Create plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create above chart with more formatting
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
# Axis lables and title
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
# Add legend
plt.legend()
```
