## 5.1.7 Scatter Plots

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Make a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Alternate
plt.scatter(x_axis, y_axis)
```

```python
# Above chart with formatting
plt.scatter(y_axis, x_axis, c='r', label='Chicago')
# Invert y axis
plt.gca().invert_yaxis()
# Axis labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
# Limits
plt.xlim(0, 50)
plt.legend()
```

### Bubble Charts

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make the bubbles larger by multiplying values
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data * 3)
```

```python
# Define dot size by new list of multiplied values
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Same as above using list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

### Scatter plots using object oriented

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create chart with formatting
fig, ax = plt.subplots(figsize=(15, 7.5))
# Plot axes, make points 5 times bigger, color sky blue, black outline, translucency
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c='skyblue', alpha=0.5,
           edgecolors='k', linewidth=2, label='Boston')
# Invert y-axis
fig.gca().invert_yaxis()
# Labels
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
# Limits
ax.set_xlim(0, 50)
# Add legend
ax.legend()
```

## 5.1.8 Pie Charts

```python
# Create pie chart using MATLAB
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Set size of chart
plt.subplots(figsize=(8, 8))
# Create chart with exploded slice and percentage values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%.1f%%')
plt.show()
```

```python
# Same as above with colors added
# Colors list
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Explode values tuple
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Size chart
plt.subplots(figsize = (8, 8))
# Create Chart with formatting
plt.pie(y_axis, 
        labels = x_axis, 
        colors = colors, 
        explode = explode_values, 
        autopct = '%.1f%%')

plt.show()
```
